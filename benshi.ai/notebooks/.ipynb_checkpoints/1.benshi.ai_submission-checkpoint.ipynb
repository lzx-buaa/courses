```python
all_products =list(train_instances.product_code.unique())
all_products
```

```python
def plot_products(dataframe: pd.DataFrame, product_code: str) -> None:
    plot_df = dataframe[dataframe.product_code==product_code]
    plot_df = plot_df[['stock_distributed']]
    plt.figure(figsize=(20,6))
    ax = plt.subplot(111)
    ax.bar(plot_df.index, plot_df['stock_distributed'], width=10)
    ax.xaxis_date()
    plt.title(f"Product - {product_code}")
    plt.grid()
    plt.show()
    return plot_df
```

```python
for p in all_products:
    plot_products(train_instances, p)
```

## Dataset Preparation

```python
from sklearn.preprocessing import LabelEncoder
```

```python
site_encoder = LabelEncoder()
product_encoder = LabelEncoder()
#train_instances.drop(columns=['timestamp'], inplace=True)
train_instances['site_code_encoded'] = site_encoder.fit_transform(train_instances['site_code'])
train_instances['product_code_encoded'] = product_encoder.fit_transform(train_instances['product_code'])
train_instances.tail()
```

```python
final_dataset=train_instances[['site_code_encoded', 'product_code_encoded', 'stock_distributed']]
final_dataset.tail()
```

```python
train_size = 0.8
train = final_dataset[:int(train_size*(len(final_dataset)))]
valid = final_dataset[int(train_size*(len(final_dataset))):]
```

## Approach 1 - Baseline - ARIMA - Naively Predict Considering Unique Data in All Sites

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
endog_list = train['stock_distributed']
exog_list = train[['site_code_encoded', 'product_code_encoded']]
dates_list = train.index
len(dates_list), len(endog_list), len(exog_list)
```

```python
model = ARIMA(endog=endog_list, 
              exog=exog_list)
model_fit = model.fit()
```

```python
exog_valid = valid[['site_code_encoded', 'product_code_encoded']]
exog_valid
```

```python
preds = model_fit.forecast(exog=exog_valid, steps=len(valid))
preds
```
