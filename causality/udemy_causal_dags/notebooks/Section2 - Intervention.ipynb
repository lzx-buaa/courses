```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
mu =0
sigma = 0.01
n_samples = 10000
e_x = np.random.normal(mu,sigma, n_samples)
e_y = np.random.normal(mu,sigma, n_samples)
e_z = np.random.normal(mu,sigma, n_samples)
```

```python
fig, ax = plt.subplots(figsize=(12,6))
sns.kdeplot(e_x, color='b', shade=True, Label='e_x') 
sns.kdeplot(e_y, color='y', shade=True, Label='e_y')
sns.kdeplot(e_z, color='c', shade=True, Label='e_z')

ax.legend()
```

```python
intervention = 1.0
z = 1*(e_z >0)
x = 1*(z + e_x >0.5)
y = 1*(x + z + e_y >2)
y_dox = 1*(intervention + z + e_y >2)
```

```python
z.mean(), x.mean(), y.mean(), y_dox.mean()
```

```python
fig, ax = plt.subplots(figsize=(12,6))
sns.kdeplot(x, color='b', shade=True, Label='x') 
sns.kdeplot(z, color='c', shade=True, Label='z')
sns.kdeplot(y, color='y', shade=True, Label='y')
sns.kdeplot(y_dox, color='r', shade=True, Label='y_dox')

ax.legend()
```
